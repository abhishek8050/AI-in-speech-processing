```python
#A1

import librosa
import librosa.display
import matplotlib.pyplot as plt
import numpy as np
import IPython.display as ipd

y, sr = librosa.load('abhishek.wav')
plt.figure(figsize=(8, 4))
librosa.display.waveshow(y)
plt.title('Speech Signal Waveform')
plt.show()
```

```python
#A2

print("signal samples:",len(y),"samples")
print("Sampling rate:",sr, "Hz")
print("Magnitude range:",np.min(y),"to", np.max(y))
```

```python
#A3

start = int(1.5 * sr)
end = int(3.5 * sr)
seg = y[start:end]
librosa.display.waveshow(seg)
ipd.Audio(seg, rate=sr)
```

```python
# A4

seg_1_s = int(1 * sr)
seg_1_e = int(2 * sr)
seg_2_s = int(2 * sr)
seg_2_e = int(3.5 * sr)
seg_1 = y[seg_1_s:seg_1_e]
seg_2 = y[seg_2_s:seg_2_e]
```

```python

print("Segment 1:")

plt.figure(figsize=(8, 4))
librosa.display.waveshow(seg_1)
plt.title('Segment-1')
plt.show()
ipd.Audio(seg_1, rate=sr)
```

```python
print("Segment 2:")
plt.figure(figsize=(8, 4))
librosa.display.waveshow(seg_2)
plt.title('Segment-2')
plt.show()
ipd.Audio(seg_2, rate=sr)
```
