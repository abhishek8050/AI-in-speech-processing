```python
import librosa
import librosa.display
import matplotlib.pyplot as plt
import numpy as np
import IPython.display as ipd

audio_signal, sample_rate = librosa.load("abhishek.wav")
first_derivative = np.diff(audio_signal)
first_derivative /= np.max(np.abs(first_derivative)) 
print("Playing First Derivative Signal:")
plt.figure(figsize=(10, 5))
librosa.display.waveshow(first_derivative, sr=sample_rate, color='blue')
plt.title('FIRST DERIVATIVE')
plt.xlabel('Time(s)')
plt.ylabel('Amplitude')
plt.grid(True)
plt.tight_layout()
plt.show()
ipd.Audio(first_derivative, rate=sample_rate)
```

```python
plt.figure(figsize=(10, 5))
librosa.display.waveshow(audio_signal, sr=sample_rate, color='blue')
plt.title('Waveform')
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
plt.grid(True)
plt.tight_layout()
plt.show()
ipd.Audio(audio_signal, rate=sample_rate)
```

```python
new_zero_crossings = np.where(np.diff(np.sign(first_derivative)))[0]
new_zero_crossing_diffs = np.diff(new_zero_crossings)
new_threshold = 1000

new_speech_intervals = new_zero_crossing_diffs[new_zero_crossing_diffs > new_threshold]
new_silence_intervals = new_zero_crossing_diffs[new_zero_crossing_diffs <= new_threshold]

new_avg_speech_interval_length = np.mean(new_speech_intervals)
new_avg_silence_interval_length = np.mean(new_silence_intervals)

print("Average length between consecutive zero crossings in speech regions:", new_avg_speech_interval_length)
print("Average length between consecutive zero crossings in silence regions:", new_avg_silence_interval_length)

plt.figure(figsize=(10, 5))
plt.plot(new_zero_crossing_diffs, label='All intervals', color='green')
plt.plot(np.arange(len(new_speech_intervals)), new_speech_intervals, 'ro', label='Speech intervals', color='blue')
plt.plot(np.arange(len(new_speech_intervals), len(new_speech_intervals) + len(new_silence_intervals)), new_silence_intervals, 'bo', label='Silence intervals', color='black')
plt.title('Pattern of Zero Crossings')
plt.xlabel('Interval Index')
plt.ylabel('Difference between Consecutive Zero Crossings')
plt.legend()
plt.show()

print("Pattern of Zero Crossings:")
print("All intervals:", new_zero_crossing_diffs)
print("Speech intervals:", new_speech_intervals)
print("Silence intervals:", new_silence_intervals)
```

```python
files_1 = ['kohli.wav', 'abd.wav', 'raina.wav', 'rahul.wav', 'dhoni.wav']
files_2 =['kohli.mp3', 'abd.mp3', 'raina.mp3', 'rahul.mp3', 'dhoni.mp3']
words = ['kohli', 'abd', 'raina', 'rahul', 'dhoni']
abhi_word_lengths = []
hari_word_lengths = []

for word_file in files_1:
    signal, sr = librosa.load(word_file, sr=None)
    length_seconds = len(signal) / sr
    abhi_word_lengths.append(length_seconds)

for word_file in files_2:
    signal, sr = librosa.load(word_file, sr=None)
    length_seconds = len(signal) / sr
    hari_word_lengths.append(length_seconds)

print("Lengths of the spoken words MINE:", abhi_word_lengths)
print("Lengths of the spoken words TeamMate:", hari_word_lengths)

bar_width = 0.35
index = np.arange(len(words))
plt.figure(figsize=(15, 5))
plt.bar(index - bar_width/2, abhi_word_lengths, bar_width, label='My Words', color='blue')
plt.bar(index + bar_width/2, hari_word_lengths, bar_width, label="Teammate's Words", color='black')
plt.xlabel('Words')
plt.ylabel('Length (seconds)')
plt.title('Comparison of Spoken Words Length')
plt.xticks(index, words)
plt.legend()

plt.show()
```
